### Удаление аутлаеров, нормирование распределений

На некоторых кривых есть выбросы и разброс в распределении. Обрежем по квантилям:

```python
# определим квантили для ограничения параметров
q1 = 0.001
q2 = 0.999

def fix_outliers(col, q1, q2):
    q_min = col.quantile(q1)
    q_max = col.quantile(q2)
    col[(col < q_min)] = q_min
    col[(col > q_max)] = q_max
```

```python
%%time
for col in list(timelog.columns)[1:-1]:
    fix_outliers(timelog[col], q1, q2)
```

### Вычисление более точного времени события по таймлогу

В таблице с осложнениями таймстемп 1 день, попробуем восстановить более точное время события по глубине нахождения инструмента `md_csg` в момент проблемы:
- Найдем интервал в 24 для нужной скважины - (`day`, `next_day`)
- Найдем близкое значение глубины к событию `depth`

```python
%%time
incident_file = 'incidents_1T.csv'
# !!! медленно
try:
    incidents = pd.read_csv(incident_file, parse_dates=['Time'])
except:
    incidents['true_time'] = '2000-01-01 00:00:00'
    for index in tqdm(range(len(incidents))):
        day = incidents.loc[index, 'Time']
        depth = incidents.loc[index, 'depth']
        well = incidents.loc[index, 'uidwellbore']
        next_day = day + timedelta(days=1)
        if well in wells:
            try:
                incidents.loc[index, 'true_time'] = timelog[(timelog['Time'] >= day) & (timelog['Time'] < next_day) & \
                    (timelog['uidwellbore'] == well) & (abs(timelog['mdCsg'] - depth) < 10)]['Time'].min()
            except:
                pass
    incidents.to_csv(incident_file, index=False)
```

## Модель предсказания инцидентов при бурении

### Обучение по всем таргетам лучшего класса модели (CatBoostClassifier + optuna)

Ищем лучшие параметры с помощью `optuna` и сохраняем модели для всех таргетов в словарь:

```python
def optimize_models(x_train, y_train, x_test, y_test, trials=100, timeout=600, ending=''):
    models = {}
    parameters = {}
    for incident in tqdm(incident_list):
        print('Optuna search for',incident)
        try:
            cbc_cycle = CatBoostClassifier()
            cbc_cycle.load_model('./models/'+incident+ending+'.cbm',
                                 format="cbm")
            models[incident] = cbc_cycle
            with open('./models/params_'+incident+ending+'.json','r') as f:
                parameters[incident] = json.load(f)
        except:
            def objective(trial):
                params = {
                    'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 0.2),
                    'depth': trial.suggest_int('depth', 4, 10),
                    'l2_leaf_reg': trial.suggest_loguniform('l2_leaf_reg', 1e-5, 10),
                    'border_count': trial.suggest_int('border_count', 32, 255),
                    'scale_pos_weight': trial.suggest_int('scale_pos_weight', 5, 15),
                    'task_type': 'GPU',
                    'loss_function' : 'Logloss',
                    'eval_metric': 'F1',
                    'verbose': False,
                    'random_state': SEED,
                    'thread_count': -1,
                    'iterations' : 1000,
                }

                # k-fold cross-validation
                kf = KFold(n_splits=5, shuffle=True, random_state=SEED)
                scores = []

                # модель и скор для каждого фолда
                for train_idx, val_idx in kf.split(x_train):
                    x_optuna_train, y_optuna_train = x_train.iloc[
                        train_idx], y_train[incident].iloc[train_idx]
                    x_optuna_val, y_optuna_val = x_train.iloc[val_idx], y_train[incident].iloc[val_idx]
                    model = CatBoostClassifier(**params)
                    model.fit(x_optuna_train, y_optuna_train,
                              eval_set=(x_optuna_val, y_optuna_val), early_stopping_rounds=50, verbose=False)
                    score = model.best_score_['validation']['F1']
                    scores.append(score)
                return sum(scores)/len(scores)

            # поиск параметров
            study = optuna.create_study(direction='maximize', study_name=incident)
            study.optimize(objective, n_trials=trials, timeout=timeout)
            parameters[incident] = study.best_params
            with open('./models/params_'+incident+ending+'.json','w') as f:
                json.dump(study.best_params, f, indent='\t')
            cbc_cycle = CatBoostClassifier(random_state=SEED,
                        task_type='GPU',
                        iterations = 1000,
                        **study.best_params)
            # обучение на подобранных параметрах
            cbc_cycle.fit(x_train, y_train[incident],
                          eval_set=(x_test, y_test[incident]), verbose=False)
            # сохранение в файл
            cbc_cycle.save_model('./models/'+incident+ending+'.cbm',
                                 format="cbm",
                                 export_parameters=None,
                                 pool=None)
            models[incident] = cbc_cycle
    return models, parameters
```

```python
%%time
models, parameters = optimize_models(x_train, y_train, x_test, y_test, trials=100, timeout=3600, ending='_1T')
models
```
